**Faktor demografis apa yang paling mempengaruhi keputusan subscription?**

```python
# CR job vs subscription
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

job_conv = (
    df_bank_cleaned
    .groupby('job')['y']
    .agg(['count','sum'])
    .reset_index()
)

job_conv['conversion_rate'] = job_conv['sum'] / job_conv['count']
job_conv = job_conv.sort_values(by='conversion_rate', ascending=False)

job_conv
```

```python
# Visualisasi job vs subscription
plt.figure(figsize=(10,5))
sns.barplot(data=job_conv, x='job', y='conversion_rate')
plt.xticks(rotation=45)
plt.title("Conversion Rate by Job")
plt.ylabel("Conversion Rate")
plt.show()
```

INSIGHTS:

Conversion rate berbeda signifikan berdasarkan jenis pekerjaan. Segmen student dan retired memiliki tingkat konversi tertinggi, sehingga menjadi target paling potensial untuk campaign deposito. Sebaliknya blue-collar dan services menunjukkan conversion rate terendah sehingga kelompok ini memerlukan pendekatan strategi yang berbeda. Hal ini menunjukkan bahwa variabel job merupakan fitur yang sangat informatif dan relevan untuk strategi targeting bisnis maupun untuk dimasukkan dalam model prediksi.

```python
# CR marital vs subscription
marital_conv = (
    df_bank_cleaned
    .groupby('marital')['y']
    .agg(['count','sum'])
    .reset_index()
)

marital_conv['conversion_rate'] = marital_conv['sum'] / marital_conv['count']
marital_conv = marital_conv.sort_values(by='conversion_rate', ascending=False)

marital_conv
```

```python
# Visualisasi marital vs subscription
plt.figure(figsize=(6,4))
sns.barplot(data=marital_conv, x='marital', y='conversion_rate')
plt.title("Conversion Rate by Marital Status")
plt.ylabel("Conversion Rate")
plt.show()
```

INSIGHTS:

Status pernikahan menunjukkan bahwa nasabah single memiliki tingkat konversi lebih tinggi dibanding divorced dan married. Hal ini mengindikasikan adanya perbedaan perilaku finansial berdasarkan kondisi pernikahan. Kategori unknown muncul karena para nasabah tidak memberikan data.

```python
# CR age group vs subscription
# Buat age group
df_bank_cleaned['age_group'] = pd.cut(
    df_bank_cleaned['age'],
    bins=[17,30,40,50,60,100],
    labels=['18-30','31-40','41-50','51-60','60+']
)
```

```python
# CR age group vs subscription
age_conv = (
    df_bank_cleaned
    .groupby('age_group')['y']
    .agg(['count','sum'])
    .reset_index()
)

age_conv['conversion_rate'] = age_conv['sum'] / age_conv['count']
age_conv = age_conv.sort_values(by='conversion_rate', ascending=False)

age_conv
```

```python
# Visualisasi age group vs subscription
plt.figure(figsize=(8,5))
sns.barplot(data=age_conv, x='age_group', y='conversion_rate')
plt.title("Conversion Rate by Age Group")
plt.ylabel("Conversion Rate")
plt.show()
```

INSIGHTS:

Data menunjukkan bahwa kelompok usia >60 tahun memiliki tingkat konversi tertinggi dibandingkan kelompok usia lainnya. Hal ini sejalan dengan karakteristik produk deposito yang bersifat rendah risiko dan stabil sehingga lebih menarik bagi nasabah usia senior yang umumnya memiliki preferensi terhadap instrumen investasi yang aman. Insight ini mengindikasikan bahwa faktor usia berperan dalam menentukan strategi segmentasi dan targeting campaign.

**Apakah jumlah kontak yang terlalu sering justru menurunkan peluang konversi?**

```python
# Buat kategori campaign
df_bank_cleaned['campaign_group'] = pd.cut(
    df_bank_cleaned['campaign'],
    bins=[0,2,5,10,100],
    labels=['1-2','3-5','6-10','10+']
)
```

```python
# Hitung CR per group
campaign_group_conv = (
    df_bank_cleaned
    .groupby('campaign_group')['y']
    .agg(['count','sum'])
    .reset_index()
)

campaign_group_conv['conversion_rate'] = (
    campaign_group_conv['sum'] / campaign_group_conv['count']
)

campaign_group_conv
```

```python
# Visualisasi CR per group
plt.figure(figsize=(8,5))
sns.barplot(data=campaign_group_conv, x='campaign_group', y='conversion_rate')
plt.title("Conversion Rate by Campaign Group")
plt.ylabel("Conversion Rate")
plt.show()
```



INSIGHTS:

Data menunjukkan bahwa efektivitas campaign tertinggi terjadi pada 1–2 kali kontak. Setelah melewati jumlah tersebut peluang konversi menurun secara konsisten yang mengindikasikan adanya diminishing return dari upaya follow up yang berlebihan.


**Bulan apa yang memiliki tingkat subscription tertinggi?**

```python
# Hitung CR per bulan
month_conv = (
    df_bank_cleaned
    .groupby('month')['y']
    .agg(['count','sum'])
    .reset_index()
)

month_conv['conversion_rate'] = month_conv['sum'] / month_conv['count']

# Urutkan dari tertinggi
month_conv = month_conv.sort_values(by='conversion_rate', ascending=False)

month_conv
```

```python
# Visualisasi CR pe bulan
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(data=month_conv, x='month', y='conversion_rate')

plt.title("Conversion Rate by Month")
plt.ylabel("Conversion Rate")
plt.xlabel("Month")

plt.show()
```

INSIGHTS:

Analisis menunjukkan bahwa tingkat subscription berbeda antar bulan. Beberapa bulan seperti March dan December memiliki conversion rate tertinggi yang mengindikasikan adanya faktor musiman atau kondisi ekonomi tertentu yang meningkatkan peluang keberhasilan campaign.

**Apakah hasil campaign sebelumnya meningkatkan peluang keberhasilan campaign berikutnya?**

```python
# Hitung CR previous = 0 vs previous > 0
# Buat group kategori
df_bank_cleaned['previous_group'] = df_bank_cleaned['previous'].apply(
    lambda x: '0 (Never Contacted)' if x == 0 else '>0 (Previously Contacted)'
)
```

```python
# Hitung CR
previous_conv = (
    df_bank_cleaned
    .groupby('previous_group')['y']
    .agg(['count','sum'])
    .reset_index()
)

previous_conv['conversion_rate'] = (
    previous_conv['sum'] / previous_conv['count']
)

previous_conv
```

```python
# Visualisasi CR previous = 0 vs previous > 0
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.barplot(data=previous_conv, x='previous_group', y='conversion_rate')

plt.title("Conversion Rate: Never vs Previously Contacted")
plt.ylabel("Conversion Rate")
plt.xlabel("Previous Contact Status")

plt.show()
```



INSIGHTS:

Analisis menunjukkan bahwa nasabah dengan riwayat campaign sebelumnya yang berhasil memiliki tingkat conversion rate jauh lebih tinggi dibandingkan dengan nasabah yang belum pernah mengikuti campaign sebelumnya. Hal ini mengindikasikan bahwa pengalaman pada campaign sebelumnya meningkatkan peluang keberhasilan pada campaign berikutnya.
