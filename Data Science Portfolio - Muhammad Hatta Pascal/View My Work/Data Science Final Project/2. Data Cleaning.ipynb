```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read Dataset
df_bank = pd.read_csv("/content/bank-additional-full.csv", sep=";")
df_bank
```

```python
# Check unknown (unknown diperlakukan sebagai kategori tersendiri)
for col in df_bank.select_dtypes(include='object').columns:
    print(col, (df_bank[col] == 'unknown').sum())
```

```python
df_bank.describe()
```

```python
# Handle pdays 999
df_bank['previously_contacted'] = df_bank['pdays'].apply(lambda x: 0 if x == 999 else 1)
```

```python
# Check
df_bank['previously_contacted'].value_counts()
```

## Cleaning

```python
# Check missing values
df_bank.isnull().sum()
```

```python
df_bank.isna().sum().sort_values(ascending=False)
```

```python
# Check duplikat
df_bank.duplicated().sum()
```

```python
# Drop duplikat
df_bank = df_bank.drop_duplicates()
```

```python
# Check
df_bank.duplicated().sum()
```

```python
# Drop kolom tidak relevan karena mengganggu model nantinya
df_bank = df_bank.drop("duration", axis=1)
df_bank
```

```python
# Check
df_bank.columns
```

```python
# Ubah y jadi numerik (encode)
df_bank['y'] = df_bank['y'].map({'no':0, 'yes':1})
```

```python
df_bank['y'].value_counts(normalize=True)
```

```python
# Handle 999 jadi -1 (belum pernah di contact), kalau 1, 2, 3 pernah
df_bank["pdays"] = df_bank["pdays"].replace(999, -1)
```

```python
# Cleaned df
df_bank.to_csv('df_bank_cleaned.csv', index=False)
```
