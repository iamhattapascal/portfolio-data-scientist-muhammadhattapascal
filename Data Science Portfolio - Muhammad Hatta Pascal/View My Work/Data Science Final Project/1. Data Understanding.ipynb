## Data Understanding

Sumber dataset : https://uci-ics-mlr-prod.aws.uci.edu/dataset/222/bank+marketing

# Tujuan Analisis :    
- Menciptakan model prediksi untuk mengidentifikasi nasabah yang memiliki kemungkinan untuk berlangganan sehingga bank dapat meningkatkan efektivitas campaign marketing dan mengurangi biaya operasional

Demografi :    
- Age = usia customer
- Job = Pekerjaan customer
- Marital = Status pernikahan
- Education = Pendidikan customer

Status finansial :    
- Default = identifikasi kredit macet customer
- Balance = saldo customer
- Housing = kredit KPR customer
- Loan = pinjaman nasabah

Campaign feature :    
- Contact
- Day
- Month
- Duration
- Campaign
- Previous
- Poutcome

Indikator ekonomi :    
- emp.var.rate = Employment variation rate
- cons.price.idx = Consumer price index
- cons.conf.idx =  Consumer confidence index
- euribor3m = Suku bunga 3 bulan
- nr.employed = Jumlah tenaga kerja

Fitur target variable :    
- y = status subscription yes or no

Metriks performa yang digunakan untuk mengukur efektivitas campaign
- Accuracy
- Precision
- Recall
- F1-Score
- Confusion Matrix
- ROC-AUC

Pertanyaan bisnis yang muncul dari dataset :    
- Demografi apa yang mempengaruhi subscription
- Apakah hasil campaign sebelumnya meningkatkan peluang keberhasilan campaign berikutnya
- Apakah jumlah kontak yang terlalu sering justru menurunkan peluang konversi
- Bulan apa yang memiliki tingkat subscription tertinggi
